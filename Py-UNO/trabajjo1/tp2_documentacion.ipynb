## Resumen de Operaciones
1. **GET**: Recuperación de información.
2. **POST**: Creación de registros.
3. **PUT**: Actualización estructural completa.
4. **PATCH**: Modificación parcial de campos.
5. **DELETE**: Eliminación definitiva de registros.

```python
import requests
import json

BASE_URL = "https://restful-booker.herokuapp.com/booking"
```

### 1. Método GET - Lectura de Datos

```python
try:
    response = requests.get(f"{BASE_URL}/1")
    response.raise_for_status()
    print("Estado: Operación exitosa")
    print(json.dumps(response.json(), indent=4))
except requests.exceptions.RequestException as e:
    print(f"Error en la petición: {e}")
```

### 2. Método POST - Creación de Recurso (Con Autenticación)

```python
auth_data = {
    "username": "admin",
    "password": "password123"
}

try:
    # Obtención de Token
    auth_res = requests.post("https://restful-booker.herokuapp.com/auth", json=auth_data)
    token = auth_res.json()["token"]
    
    # Creación de Reserva
    reserva = {
        "firstname": "Ezequiel",
        "lastname": "Enrico Areco",
        "totalprice": 120,
        "depositpaid": True,
        "bookingdates": {"checkin": "2026-04-01", "checkout": "2026-04-05"}
    }
    
    headers = {"Cookie": f"token={token}"}
    response = requests.post(BASE_URL, json=reserva, headers=headers)
    print(f"Código de Estado: {response.status_code}")
    print(json.dumps(response.json(), indent=4))
except Exception as e:
    print(f"Error de ejecución: {e}")
```
